```python
tech_companies = ['Google', 'Apple', 'Facebook', 'Uber', 'Intel']
```

```python
list_len = len(tech_companies)

list_len
```

```python
count = 0

while count < list_len:
    print(tech_companies[count])
    
    count = count + 1
```

```python
inc_count = 0
dec_count = 10

while inc_count > 5 or dec_count > 5:
    print(inc_count, dec_count)
    
    inc_count = inc_count + 1
    dec_count = dec_count - 2

print('Final count values ', inc_count, dec_count)    
```

```python
num = 8

while num < 10:
    num = int(input('Enter a number: '))
    
    print('You entered: ', num)
    
print('\nFinal number which caused us to break out of the loop', num)
```

```python
num = 8

while True:
    num = int(input('Enter a number: '))
    
    print('You entered: ', num)
    if num % 5 == 0:
        break
    
print('\nFinal number which caused us to break out of the loop', num)    
```

```python
num = 8

while True:
    num = int(input('Enter a number: '))
    
    if num % 2 == 0:
        continue
    
    print('\nYou entered: ', num)
    
    if num % 5 == 0:
        break
    
print('\nFinal number which caused us to break out of the loop', num)    
```
